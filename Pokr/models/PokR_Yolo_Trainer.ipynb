```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import torch



path = r"C:\Users\thenn\Documents\Collection\FUN\PokR Mini\img\many.jpg"
im = cv2.imread(path)
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
im_resised = cv2.resize(im,(640,640))
print(type(im))
im

results=model.predict(source=im_resised,conf=0.005)


out = results[0].boxes.data

results_tensor = out
class_names = ['10C', '10D', '10H', '10S', '2C', '2D', '2H', '2S', '3C', '3D', '3H', '3S', '4C', '4D', '4H', '4S', '5C', '5D', '5H', '5S', '6C', '6D', '6H', '6S', '7C', '7D', '7H', '7S', '8C', '8D', '8H', '8S', '9C', '9D', '9H', '9S', 'AC', 'AD', 'AH', 'AS', 'JC', 'JD', 'JH', 'JS', 'KC', 'KD', 'KH', 'KS', 'QC', 'QD', 'QH', 'QS']

# Load the original image
original_image = im_resised

# Create a plot with two subplots: one for the original image and one for the image with bounding boxes
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Display the original image
axes[0].imshow(original_image)
axes[0].set_title("Original Image")
axes[0].axis("off")

# Display the image with bounding boxes
axes[1].imshow(original_image)
axes[1].set_title("Image with Bounding Boxes")
axes[1].axis("off")

# Plot bounding boxes and class names on the second subplot
for box in results_tensor:
    x_min, y_min, x_max, y_max, confidence, class_id = box.tolist()
    print(box.tolist())
    class_id = int(class_id)
    class_name = class_names[class_id]
    
    # Draw the bounding box
    rect = patches.Rectangle(
        (x_min, y_min), x_max - x_min, y_max - y_min, 
        linewidth=2, edgecolor='red', facecolor='none'
    )
    axes[1].add_patch(rect)
    
    # Add class name and confidence score as text
    axes[1].text(
        x_min, y_min - 5, f"{class_name} ({confidence:.2f})", 
        color='red', fontsize=10, fontweight='bold'
    )

# Show the plot
plt.tight_layout()
plt.show()
```
